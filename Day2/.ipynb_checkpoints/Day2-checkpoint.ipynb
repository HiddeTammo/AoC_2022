```python
#Input
f = open('input.txt', 'r')
x = f.read().splitlines()
x
```

```python
#Parameters
l = len(x)
print(l)
score = 0

#Outcome
lose = 0
draw = 3
win  = 6

for i in range(l):
    
    #Split result 
    r = x[i].split(' ')
    
    if r[0] == "A":
        r[0] = 1
    elif r[0] == "B":
        r[0] = 2
    else:
        r[0] = 3
    
    if r[1] == "X":
        r[1] = 1
    elif r[1] == "Y":
        r[1] = 2
    else:
        r[1] = 3
    
    if r[0] == r[1]:
        score += draw + r[1]
    elif r[0] < r[1]:
        score += win + r[1]
    elif r[0] == 3 and r[1] == 1:
        score += win + r[1]
    else:
        score += r[1]
    


        
print(score)
```
