# Part 2

```python
#Input
f = open('input.txt', 'r')
x = f.read().splitlines()
x
```

```python
#parameters
length = len(x)
ans = []

for i in range(0, length, 3):
    
    #Input from list
    e1, e2, e3 = x[i], x[i+1], x[i+2]
    
    
    #Conditions for while loop
   
    found = False
    
    while found != True:
        
        #sL = String length of each elf
        stringLen = [len(e1), len(e2), len(e3)]
        
        print(e1)
        print(e2)
        print(e3)
        
        #Pops current letter from elf 1 from all rucksacks
        c = e1[0]
        e1 = e1.replace(c,"")
        e2 = e2.replace(c,"")
        e3 = e3.replace(c,"")
    
        #Found character, save it and break while loop
        if len(e1) != stringLen[0] and len(e2) != stringLen[1] and len(e3) != stringLen[2]:
            print(c)
            ans.append(c)
            found = True

ans
```

```python
#Score
ansLen = len(ans)
ans2 = []
for i in range(ansLen):
    if ans[i].islower():
        ans2.append(ord(ans[i]) - 96)
    if ans[i].isupper():
        ans2.append(ord(ans[i]) - 64+26)

y = sum(ans2)
y
```
